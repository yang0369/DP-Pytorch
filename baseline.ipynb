```python
#分析函数
def analysis(valid_y, valid_y_pre,thresh=0.5,top=0.1,print_thresh=False,step=50):
    '''
    阈值减低到很低,来提高召回率 的尝试
    '''
    scores={}
    for thresh in [x/1000 for x in list(range(50,601,step))]:
        valid_pre_bin = pd.Series(valid_y_pre).apply(lambda x: 1 if x > thresh else 0)
        m = confusion_matrix(valid_y, valid_pre_bin)
        m = pd.DataFrame(m, index=['真0', '真1'], columns=['预测0', '预测1'])
        #print('a:',classification_report(valid_y, valid_pre_bin))
        score=f1_score(valid_y,valid_pre_bin)
        if print_thresh:
            print('thresh:',thresh,'f1_score %.4f'%score)
        scores[score]=thresh
        #print('混淆矩阵\n', m)
    auc = roc_auc_score(valid_y, valid_y_pre)
    f1_final=max(scores)
    
    valid_pre_bin = pd.Series(valid_y_pre).apply(lambda x: 1 if x > scores[f1_final] else 0)
    rc=recall_score(valid_y, valid_pre_bin)
    ac=accuracy_score(valid_y, valid_pre_bin)
    if print_thresh:
        print('Final AUC:%.4f,F1:%.4f@thresh%s ,sum(AUC+F1):%.4f with recall%.4f,precisiion%.4f' % (auc,f1_final,scores[f1_final],auc+f1_final,rc,ac) )
    
    #     fpr, tpr, thres = roc_curve(valid_y, valid_y_pre)
    #     print('\nKS: ', max(tpr-fpr))
    #     fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
    #     ax1.plot(fpr, tpr)
    #     ax1.set_title('roc')
    #     ax2.plot(thres[1:], tpr[1:])
    #     ax2.plot(thres[1:], fpr[1:])
    #     ax2.plot(thres[1:], tpr[1:]-fpr[1:])
    #     ax2.legend(['tpr', 'fpr', 'tpr-fpr'])
    #     ax2.set_title('ks')
    #     ax2.invert_xaxis()
    
    sorted_valid_y = pd.Series(valid_y_pre).sort_values(ascending=False)
    if print_thresh:
        print('正样本总量', sum(valid_y))
    n=sum(valid_y)
    nums=[100]+[500*2**i for i in range(5)]+[10000]
    ans=[]
    for num in nums:
        tmp=('top %s 命中数 '%num,sum(valid_y[list(sorted_valid_y[:num].index.values)]),'命中率','%.4f'%(sum(valid_y[list(sorted_valid_y[:num].index.values)])/num),'召回率','%.4f'%(sum(valid_y[list(sorted_valid_y[:num].index.values)])/n))
        #ans.append(tmp) #暂时去掉
    num=int(len(valid_pre_bin)*0.1)
    tmp=('top 0.1 %s 命中数 '%num,sum(valid_y[list(sorted_valid_y[:num].index.values)]),'命中率','%.4f'%(sum(valid_y[list(sorted_valid_y[:num].index.values)])/num),'召回率','%.4f'%(sum(valid_y[list(sorted_valid_y[:num].index.values)])/n),)
    ans.append(tmp)    
    
    return auc+f1_final,[auc,f1_final,auc+f1_final,scores[f1_final]], 'Final AUC:%.4f,F1:%.4f@thresh%s ,\n sum(AUC+F1):%.4f \n with recall%.4f,precisiion%.4f'%(auc,f1_final,scores[f1_final],auc+f1_final,rc,ac)  

#analysis(results.label,results.pre,)    
#analysis(valid_y.reset_index(drop=True), valid_y_pre)
```

```python
#feat2=['lower_30_10_rate', 'deal_60_10_count_rate', 'per_deal_90_6_18_cnt', 'deal_tract50_90_out_cnt', 'account_sum', 'open_i_acct', 'open_card_num', 'deal_90_5_count_rate', 'open_amt', 'marital_status', 'create_sum', 'deal_90_23_6_cnt', 'balance_90_day', 'othbk_cnpy_num_90', 'change_passwd_30_cnt', 'freq_block_trade_90_day', 'per_deal_90_18_23_cnt', 'aver_15_timestamp', 'deal_mbank_cnt_60', 'deal_avrg_30_90_cnt_rate', 'trans_prov_trade_90_day', 'freq_trade_90_day', 'create_account_cnt', 'deal_avrg_30_cnt', 'cnpy_7_90_rate', 'deal_avrg_90_cnt', 'deal_avrg_30_trin_amt', 'freq_trade_30_day', 'deal_avrg_30_trout_amt', 'deal_90_6_18_cnt', 'deal_tract10_90_out_cnt', 'cnpy_num_7', 'change_card_cnt', 'freq_block_trade_30_day', 'is_cre_card', 'empty_remark_90_rate', 'pos_trans_amount_ctrl', 'per_deal_30_tract_amt', 'cancel_account_cnt', 'cancel_sum', 'self_help_trans_cnt_ctrl', 'trans_prov_trade_30_day']
#feat2+=['y']
def prepare(df,if_drop=False,if_flag=False,if_cat=False,if_na=False,if_add_feature=False):
    df=df.copy()
    df=df.fillna(0)
    if_rhfeat=False
    def add_feature(df):
        df_input = df.copy()
        df_input['cancel_create'] = df_input['cancel_sum']/df_input['create_sum']
        df_input['i_acct_ratio']=df_input['open_i_acct']/df_input['open_card_num'] #开立一类卡占总借记卡笔数
        df_input['account_ratio']=(df_input['create_sum']-df_input['cancel_sum'])/df_input['account_sum'] # 近十年开-销/正常账户数 (剩余账户)
        df_input['account_sum_ratio']=(df_input['create_sum']+df_input['cancel_sum'])/df_input['account_sum'] # 近十年开+销/正常账户数（频繁程度）
        df_input['account_i_ratio']=df_input['account_sum']/df_input['open_i_acct'] #同业账户总数/开立I类户数
        df_input['account_card_ratio']=df_input['account_sum']/df_input['open_card_num'] #同业账户总数/开立I类户数
        df_input['draw_30_avg_amt']=df_input['draw_30_amt']/df_input['draw_30_cnt'] #近30天活期账户现金支取金额均值
        df_input['dep_30_avg_amt']=df_input['dep_30_amt']/df_input['dep_30_cnt'] #近30天活期账户现金存入金额均值
        df_input['draw_30_amt_ratio']=df_input['draw_30_amt']/(df_input['dep_30_amt']+df_input['draw_30_amt']) #近30天支取金额比率
        df_input['draw_30_cnt_ratio']=df_input['draw_30_cnt']/(df_input['dep_30_cnt']+df_input['draw_30_cnt']) #近30天支取次数比率
        df_input['deal_tract50_cnt_ratio']=df_input['deal_tract50_90_out_cnt']/df_input['deal_avrg_90_cnt'] #近90天转出金额小于50元的笔数比率
        df_input['deal_tract10_cnt_ratio']=df_input['deal_tract10_90_out_cnt']/df_input['deal_avrg_90_cnt'] #近90天转出金额小于10元的笔数比率
        df_input['deal_mbank_cnt_60_30_ratio']=df_input['deal_mbank_cnt_60']/df_input['deal_avrg_30_cnt']/60 #近60天手机银行渠道交易笔数/近30天日均交易笔数
        df_input['deal_mbank_cnt_60_90_ratio']=df_input['deal_mbank_cnt_60']/df_input['deal_avrg_90_cnt']/60 #近60天手机银行渠道交易笔数/近90天日均交易笔数
        df_input['freq_trade_30_90_ratio']=df_input['freq_trade_30_day']/df_input['freq_trade_90_day'] #近30天频繁交易天数/近90天频繁交易天数
        df_input['high_freq_block_trade_30_90_ratio']=df_input['high_freq_block_trade_30_day']/df_input['high_freq_block_trade_90_day'] #近30天高频大额交易天数/近90天高频大额交易天数
        df_input['change_30_cnt']=df_input['change_passwd_30_cnt']+df_input['change_phone_30_cnt'] #近30天修改次数
        df_input['balance_30_90_day_ratio']=(df_input['balance_30_day']/30)/(df_input['balance_90_day']/90) #近30天/90天资金日平率
        df_input['trans_prov_trade_7_30_day_ratio']=(df_input['trans_prov_trade_7_day']/7)/(df_input['trans_prov_trade_30_day']/30) #跨省交易7天/9天
        df_input['trans_prov_trade_7_90_day_ratio']=(df_input['trans_prov_trade_7_day']/7)/(df_input['trans_prov_trade_90_day']/90) #跨省交易7/90天
        df_input['trans_prov_trade_30_90_day_ratio']=(df_input['trans_prov_trade_30_day']/30)/(df_input['trans_prov_trade_90_day']/90) #跨省交易30/90天
        df_input['aver_15_timestamp']=df_input['aver_15_timestamp']/(24*3600) #将间隔时间秒转换为天
        df_input['trans_cnt_ctrl_total']=df_input['self_help_trans_cnt_ctrl']+df_input['pos_trans_amount_ctrl'] #一年内账户预警总数
        return df_input
    if if_add_feature:
        df = add_feature(df)
    if if_drop:
        for c in ['rapid_trade_7_5_cnt','rapid_trade_7_30_cnt','rapid_trade_30_5_cnt','rapid_trade_30_30_cnt','high_freq_trade_30_day','high_freq_trade_90_day','balance_30_day']:
            if c in df:
                df = df.drop(c, axis=1)
    if if_flag:
        df['flag'] = (df['cnpy_7_90_rate'] >1) | (df['deal_avrg_30_90_cnt_rate'] >1) | (df['deal_90_5_count_rate'] >1) | (df['deal_60_10_count_rate'] >1) | (df['lower_30_10_rate'] >1) | (df['empty_remark_30_rate'] >1) | (df['empty_remark_90_rate'] >1)
    #     if if_rhfeat:#三类人行特征
    #         if 'account_sum' in df.columns:
    #             print('删除人行3特征')
    #             df=df.drop(['account_sum','cancel_sum','create_sum'],axis=1)
    #更换类型 
    str_cols='high_freq_trade_30_day, high_freq_trade_90_day, balance_30_day, draw_30_amt, dep_30_amt, cnpy_7_90_rate, per_deal_90_6_18_cnt, per_deal_90_18_23_cnt, deal_avrg_30_cnt, deal_avrg_90_cnt, per_deal_30_tract_amt, deal_avrg_30_trin_amt, deal_avrg_30_trout_amt, deal_avrg_30_90_cnt_rate, aver_15_timestamp, deal_90_5_count_rate, deal_60_10_count_rate, lower_30_10_rate, open_i_acct, deal_90_6_18_cnt, deal_90_23_6_cnt, deal_tract50_90_out_cnt, deal_tract10_90_out_cnt, freq_block_trade_30_day, freq_block_trade_90_day, high_freq_block_trade_30_day, high_freq_block_trade_90_day, change_card_cnt, balance_90_day, trans_prov_trade_7_day, trans_prov_trade_30_day, trans_prov_trade_90_day'.split(', ')
    str_cols_train='OPEN_AMT, IF_HOLD_FINC, DRAW_30_AMT, DEP_30_AMT, CNPY_7_90_RATE, PER_DEAL_90_6_18_CNT, PER_DEAL_90_18_23_CNT, DEAL_AVRG_30_CNT, DEAL_AVRG_90_CNT, PER_DEAL_30_TRACT_AMT, DEAL_AVRG_30_TRIN_AMT, DEAL_AVRG_30_TROUT_AMT, DEAL_AVRG_30_90_CNT_RATE, AVER_15_TIMESTAMP, DEAL_90_5_COUNT_RATE, DEAL_60_10_COUNT_RATE, LOWER_30_10_RATE, EMPTY_REMARK_30_RATE, EMPTY_REMARK_90_RATE'.split(', ')
    str_cols+=str_cols_train
    #print(len(str_cols))
    ct=0
    for c in str_cols:
        if c in df.columns:
            df[c] = df[c].astype(float)
            #print('change:',c)
            ct+=1
    cat_cols=['marital_status']
    if if_cat:
        print('类别转换')
        for c in cat_cols:
            if c in df.columns:
                df[c] = df[c].astype('category')
    if if_na:
        for c in ['t_id','marital_status','is_cre_card','if_hold_finc','y']:
            if c in df:
                rh_num = df.drop(c,axis=1)
        df['zero_ratio'] = (rh_num==0).astype(float).sum(axis=1)/rh_num.shape[1]  
        
    print('类型转换完毕,共计%s列,转float'%ct)
    if 'PASSWD_PHONE_TIME_DIFF' in df:
        df = df.drop('PASSWD_PHONE_TIME_DIFF', axis=1)
    if 'PASSWD_PHONE_TIME_DIFF'.lower() in df:
        df = df.drop('PASSWD_PHONE_TIME_DIFF'.lower(), axis=1)
    
    return df
```
